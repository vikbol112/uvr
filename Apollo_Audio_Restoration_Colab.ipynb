```python
%cd /content/Apollo
#@markdown #Inference
#@markdown For the universal model set *chunk_size* above to 19, for all other models set it to 25
input_file_path = '/content/Gyllene Tider - Ska vi älska, så ska vi älska till Buddy Holly.flac' #@param {type:"string"}
output_file_path = '/content/output.wav' #@param {type:"string"}
model = 'Custom Model' #@param ['MP3 Enhancer', 'Lew Vocal Enhancer', 'Lew Vocal Enhancer v2 (beta)', 'Lew Universal Lossy Enhancer', 'Custom Model']
chunk_size = 8 #@param {type:"slider", min:3, max:25, step:1}
overlap = 4 #@param {type:"slider", min:2, max:10, step:1}

if model == 'MP3 Enhancer':
    ckpt = '/content/Apollo/model/pytorch_model.bin'
    config = 'configs/apollo.yaml'
elif model == 'Lew Vocal Enhancer':
    ckpt = '/content/Apollo/model/apollo_model.ckpt'
    config = 'configs/apollo.yaml'
elif model == 'Lew Vocal Enhancer v2 (beta)':
    ckpt = '/content/Apollo/model/apollo_model_v2.ckpt'
    config = 'configs/config_apollo_vocal.yaml'
elif model == 'Lew Universal Lossy Enhancer':
    ckpt = '/content/Apollo/model/apollo_model_uni.ckpt'
    config = 'configs/config_apollo_uni.yaml'
elif model == 'Custom Model':
    ckpt = '/content/Apollo/model/model_apollo_0129_-21.98.ckpt'
    config = '/content/Apollo/model/config_apollo_0129_-21.98_1778847391.yaml'

# Patch inference.py to prevent KeyError if model_name is missing
!sed -i 's/conf\["model_name"\]/conf.get("model_name", "Apollo")/g' /content/Apollo/look2hear/models/base_model.py

!python inference.py \
    --in_wav '{input_file_path}' \
    --out_wav '{output_file_path}' \
    --chunk_size {chunk_size} \
    --overlap {overlap} \
    --ckpt '{ckpt}' \
    --config '{config}'
```
